## (Nashville) art in public places 

```python
endpoint7 ='https://data.nashville.gov/resource/dqkw-tj5j.json'
```

```python
response = requests.get(endpoint7)
response
```

```python
result3 = response.json()
result3_df=pd.DataFrame(result3)
result3_df
```

```python
art_df = pd.merge (result2_df, result3_df, on = "longitude", how = "left")
```
